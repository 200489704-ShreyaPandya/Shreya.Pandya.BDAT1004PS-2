```python
class Marsupial:
    def __init__(self):
        self.itemLst = []
    
    def put_in_pouch(self,item):
        self.itemLst.append(item)
        
    def pouch_contents(self):
        return self.itemLst
    
class Kangaroo(Marsupial):
    def __init__(self,dx,dy):
        self.x = dx
        self.y = dy
    
    def jump(self,dx,dy):
        self.x += dx
        self.y += dy
    
    def __str__(self):
        return "I am a kangaroo located at coordinates ({0},{1})".format(self.x,self.y)
```

```python
M = Marsupial()
M.put_in_pouch('Doll')
M.put_in_pouch('FireTruck')
M.put_in_pouch('Kitten')
M.pouch_contents()
```

```python
#Reference: https://docs.python.org/3/library/html.parser.html

from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    (h1,h2,h3,h4,h5,h6) = (False,False,False,False,False,False)
    
    def handle_starttag(self,tag,attrs):
        if tag == 'h1':
            self.h1 = True
        if tag == 'h2':
            self.h2 = True
        if tag == 'h3':
            self.h3 = True
        if tag == 'h4':
            self.h4 = True
        if tag == 'h5':
            self.h5 = True
        if tag == 'h6':
            self.h6 = True
            
    def handle_endtag(self,tag):
        if tag == 'h1':
            self.h1 = False
        if tag == 'h2':
            self.h2 = False
        if tag == 'h3':
            self.h3 = False
        if tag == 'h4':
            self.h4 = False
        if tag == 'h5':
            self.h5 = False
        if tag == 'h6':
            self.h6 = False

    def handle_data(self,data):
        if self.h1:
            print(data)
        elif self.h2:
            print("\t{}".format(data))
        elif self.h3:
            print("\t\t{}".format(data))
        elif self.h4:
            print("\t\t\t{}".format(data))
        elif self.h5:
            print("\t\t\t\t{}".format(data))
        elif self.h6:
            print("\t\t\t\t\t{}".format(data))

#Main Code

File = open('w3c.html')
FileContent = File.read()
File.close()

HeadingFormatParser = HeadingParser()
HeadingFormatParser.feed(FileContent)
```

```python
from urllib.parse import urljoin
from html.parser import HTMLParser
from urllib.request import urlopen
class Collector(HTMLParser):
    'collects hyperlink URLs into a list'

    def __init__(self, url):
        'initializes parser, the url, and a list'
        HTMLParser.__init__(self)
        self.url = url
        self.links = []

    
    def handle_starttag(self, tag, attrs):
        'collects hyperlink URLs in their absolute format'
#         print("Start tag called")
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    # construct absolute URL
                    print("-->Before ",attr[1])
                    absolute = urljoin(self.url, attr[1])
                    print("-->absolute ",absolute)
                    if absolute[:4] == 'http': # collect HTTP URLs
                        self.links.append(absolute)
                        
    def getLinks(self):
        'returns hyperlinks URLs in their absolute format'
        return self.links

```

```python
def webdir(url,depth,indent):
    if(depth<0):
        return
    resource = urlopen(url)
    collector = Collector(url)
    collector.feed(resource.read().decode())
    print("\t"*indent +url)
    for link in collector.getLinks():
        webdir(link,depth-1,indent+1)
        
# Main code

#print("--"*15+"With Depth of 2"+"--"*15)
#webdir("https://reed.cs.depaul.edu/lperkovic/csc242/test1.html",2,0)

print("--"*15+"With Depth of 2"+"--"*15)
webdir("http://reed.cs.depaul.edu/lperkovic/test1.html",2,0)
```
